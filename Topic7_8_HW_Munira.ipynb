```python
fig, (ax1, ax2) = plt.subplots(2)
fig.suptitle('Body_mass_g Variable Prediction')

scatter_plot_1 = sns.scatterplot(data = df3, x='body_mass_g', y = 'flipper_length_mm', hue = 'island', ax = ax1)
scatter_plot_2 = sns.scatterplot(data = df3, x='body_mass_g', y = 'bill_length_mm', hue = 'island', ax = ax2)

ax1.set_xlabel('Body Mass')
ax2.set_xlabel('Body Mass')
ax1.set_ylabel('Flipper Length')
ax2.set_ylabel('Bill Length')

fig.set_figwidth(10)
fig.set_figheight(8)

plt.show()
```
